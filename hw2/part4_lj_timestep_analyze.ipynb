```python
from HW2_solutions import average
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Warmup was around 1000 steps = 10 time units


```python
ke = [] 
ke_err = []
timesteps = [0.01, 0.02, 0.04, 0.1, 0.2, 0.4 ]
for timestep in timesteps: 
    df = pd.read_csv(f"data/timestep{timestep}.csv")
    k, k_err, kappa, var = average(df['KE'].values[int(10/timestep):])
    ke.append(k)
    ke_err.append(k_err)
```

```python
plt.errorbar(timesteps, ke, ke_err, marker='o', mec='k')
plt.xlabel("Time step")
plt.ylabel("Kinetic energy (LJ units)")
```
